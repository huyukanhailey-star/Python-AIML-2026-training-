```python
def gencubes(n):
    for num in range(n):
        yield num**3

gencubes(10)

for x in gencubes(10):
    print(x)
```

# map 
map(function, sequence)
The first argument is the name of a function and the second a sequence (e.g. a list). map() applies the function to all the elements of the sequence. 
It returns a new list with the elements changed by the function.

```python
def fahrenheit(T):
    return ((float(9)/5)*T + 32)

def celsius(T):
    return (float(5)/9)*(T-32)

temp = [0, 22.5, 40, 100]

fahrenheit_temps = map(fahrenheit, temp)
print(list(fahrenheit_temps))

celsius_temps = map(celsius, temp)
print(list(celsius_temps))
```

# reduce 
reduce(function, sequence) 

continually applies the function to the sequence
then returns a single value

The reduce() function - responsible for "remembering the result from the previous step, and feeding that result — together with the next new element — into your lambda for the next round."

# lambda  one line dummy function    use colon  :


```python
from functools import reduce
lst = [1,2,3,4,5,6]
sum_lst = reduce(lambda x, y: x + y, lst)  # function lambda x, y: x + y  # sequence lst
print(sum_lst)
```

```python
max_find = lambda a,b : a if a>b else b
lst = [47,49,42,55,56]

from functools import reduce
reduce(max_find,lst)
```

# filter

convenient way

```python
words = ["ruchi", 'Nikhil', 'phoebee', 'swapna']
long_words = list(filter(lambda word: len (word)>=6, words) )
print (long_words)
```
